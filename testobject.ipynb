```python
#輸出json串流,json.loads()用於將str型別的資料轉成dict
def readjson(url):
    #python終端機是Linux系統,window執行會是cp950,所以要轉utf-8
    response = json.load(open(url,encoding="utf-8"))
    '''
    print(response)#輸出
    print(type(response)) 
    print (response['494255'])
    '''
    return response
#檢查是否全是中文字符
def is_all_chinese(strs):
    for _char in strs:
        if not '\u4e00' <= _char <= '\u9fa5':
            return False
    return True
#判斷是否為需要的詞
def isneed(a,b):
    if(('n' in b or 'x' in b) and is_all_chinese(a) and 'eng' not in b and a not in punc and a not in punc2 and not a.isdigit()):
        return True
    else:
        return False
```

```python
#宣告全域變數
key_val = defaultdict(defaultdict)#宣告主要字典存放資料
L = []#宣告陣列暫存資料
punc='！”＃＄％＆’（）、＊＋，－。／：；＜＝＞？＠〔＼〕︿ˍ‘｛｜｝～ ';#標點符號(全形)+空白
punc2=string.punctuation;#標點符號(半形)+空白
sentence_count=0#句數
starttime = datetime.datetime.now()#起始時間
```

```python
#第一個檔案Posseg_List9.json
response=readjson('test.json')
for key in response.keys():
    for key2 in response.get(key):
        x = key2.split("/", 1)
        #判斷詞性為名詞與過濾詞
        if(isneed(x[0],x[1])):
            #print('%s' % (x[0]))
            L.append(x[0])
            #key_val[x[0]] = 0
    L2 = L.copy()
    for i in L:
        for j in L2:
            #print(i,j)
            #自己跟自己就不算進去
            if i!=j:
                try:
                    key_val[i][j]=key_val[i][j]+1
                except:
                    key_val[i][j]=0
    L=[]#清空
    L2=[]#清空
    #print('--'*30)
    #sentence_count=sentence_count+1
    #print('第 %d句'%(sentence_count))
#print(sentence_count)
#print(key_val)
print("end")
endtime = datetime.datetime.now()
print (endtime - starttime)
```
